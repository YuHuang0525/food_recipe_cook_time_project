```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.remote.webelement import WebElement
from selenium.webdriver.common.keys import Keys
import re
import time
import gzip
```

```python
driver2 = webdriver.Chrome(ChromeDriverManager().install())
wait2 = WebDriverWait(driver2, 30)
```

```python
def readGz(path):
    for l in gzip.open(path, 'rt',encoding='utf-8'):
        yield eval(l)
```

```python
test_recipes = []
for d in readGz("testRecipes.json.gz"):
    test_recipes.append(d)
    
test_recipes_name = []
for x in test_recipes:
    test_recipes_name.append(x['name'])

test_recipes_name_id = dict()
for r in test_recipes:
    test_recipes_name_id[r['name']] = r['recipe_id']
```

```python
# 记录名字搜出来是oops的recipe
ambiguous_recipe = []
```

```python
minutes_8000 = {}
```

```python
def search(r):
    submit = wait2.until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'body > div.fd-site > div.fd-site-wrapper > section.gk-header-wrap.Food\.com > header > div > div > div > nav:nth-child(2) > ul > li.v-nav-toggle.has-icon.gk-nav__search > a > i'))
        )
    submit.click()

    # input the search recipe
    # 找text box需要通过id去找 注意！！！
    inputElement = driver2.find_element_by_id("search-input")
    inputElement.send_keys(r)

    # hit enter after inputting the text
    inputElement.send_keys(Keys.RETURN)
```

```python
executed = 0
```

```python
for r in test_recipes_name[9500:10000]:
    try:
        
        search(r)
        
        # 这个地方注意从原来的implict wait换成这个time sleep是因为
        # implicit wait是如果找不到对应的element一直等到 多久throw exception，但是
        # 网页一旦加载 那个信息就会出来，只是不包含我们想要的有效字段，所以我们怎么判断有没有我们要的
        # 关键字都没用，因为间隔时间太短太短了，不足以让我们要的字段上传。
        time.sleep(1.5)
        
        e = driver2.find_element_by_css_selector('#searchModuleTitle')
        
        # get the entire thing of the corresponding div class
        temp = e.get_attribute('outerHTML')
        
        if '1 Result' in temp:
            driver2.implicitly_wait(2)
            
            e = driver2.find_element_by_css_selector('#gk-menu-search > div.fd-menu-inner > div.search-module-results.on > div > div > div > div > div.tile-content > div > div > div.meta-data > div.cook-time')
            # get the entire thing of the corresponding div class
            temp = e.get_attribute('outerHTML')
            # using regular expression to find the text in between
            t = re.findall('[0-9]+[\s\.]*[0-9]*\s+[hm]*', temp)[0].strip()
            temp_r_id = test_recipes_name_id[r]
            minutes_8000[temp_r_id] = t
        else:
            continue
            
        executed += 1
        print(executed)
        
    except:
        # find oops recipe
        ambiguous_recipe.append(r)

        executed += 1
        print(executed)
```

```python
%store minutes_8000
```

```python
minutes_8000_revised = dict()
for x,y in minutes_8000.items():
    temp = y.strip().split(' ')
    if 'h' in temp:
        minutes_8000_revised[x] = float(temp[0]) * 60
    elif 'm' in temp:
        minutes_8000_revised[x] = float(temp[0])
    else:
        print(x)
```

```python
minutes_8000_to_csv_list = []
for x, y in minutes_8000_revised.items():
    minutes_8000_to_csv_list.append([x,y])
```

```python
import csv
```

```python
with open('actual_minutes_8000.csv', 'w') as f:
    write = csv.writer(f)
    write.writerows(minutes_8000_to_csv_list)
```
